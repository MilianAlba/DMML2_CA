```python
from keras.preprocessing.image import ImageDataGenerator
from skimage import io
datagen = ImageDataGenerator(        
        rotation_range = 40,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        #height_shift_range = 0.2,
        brightness_range = (0.5, 1.5))
import numpy as np
import os
from PIL import Image
#image_directory = r'C:Users---train/'
#image_directory = r"C:\Users\bodwy\plantvillage dataset\color\Tomato___healthy/"
image_directory = r"C:\Users\bodwy\PlantDoc-Dataset\train\Tomato Early blight leaf/"
#SIZE = 224
SIZE = 256
dataset = []
my_images = os.listdir(image_directory)
print(my_images)
j=0
for i, image_name in enumerate(my_images):    
    #if (image_name.split('.')[1] == 'JPG'):
    if image_name.endswith('.jpg'):
        j += 1
        image = io.imread(image_directory + image_name)        
        image = Image.fromarray(image, 'RGB')        
        image = image.resize((SIZE,SIZE)) 
        dataset.append(np.array(image))
x = np.array(dataset)
print(j)
print(len(dataset))
print(x.shape)
i = 0
for batch in datagen.flow(x, batch_size=32,
                          #save_to_dir= r'C:Users---Augmented-images',
                          #save_to_dir= r"C:\Users\bodwy\plantvillage dataset\color\Tomato___healthy_Augmented/",
                          save_to_dir= r"C:\Users\bodwy\PlantDoc-Dataset\train\Tomato Early blight leaf Augmented/",
                          save_prefix='dr',
                          save_format='jpg'):    
    i += 1
    print(i)
    if i > 50:        
        break
```
